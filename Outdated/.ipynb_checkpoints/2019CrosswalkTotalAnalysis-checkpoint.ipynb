```python
import csv
from collections import OrderedDict
import arcgis
from IPython.display import display
import pandas as pd
import openpyxl
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
field_list = ['SUBTYPE','BARS','CREW_ASSIGNED','INSTALL_DATE','SHAPE_LEN']
crosswalks = pd.read_csv(r"C:\Users\Govs\CrosswalkData2019.csv")
crosswalks = crosswalks.filter(items=field_list)
crosswalks["INSTALL_DATE"] = pd.to_datetime(crosswalks['INSTALL_DATE'])

def create(df,q):
    temp = crosswalks.query("CREW_ASSIGNED == " + q).groupby("SUBTYPE").agg({'BARS':'sum','SHAPE_LEN':'sum','CREW_ASSIGNED':'count'})
    temp["Filter"] = q[1:-1]
    return temp

# Create table of 2018 crosswalks
y18 = crosswalks.query('20171001 < INSTALL_DATE < 20181001').groupby("SUBTYPE").agg({'BARS':'sum','SHAPE_LEN':'sum','CREW_ASSIGNED':'count'})
school = create(crosswalks, "'SCHOOL'")
signal = create(crosswalks, "'SIGNAL'")
cbd = create(crosswalks, "'CBD'")
other = create(crosswalks, "'OTHER'")
total = crosswalks.groupby("SUBTYPE").agg({'BARS':'sum','SHAPE_LEN':'sum','CREW_ASSIGNED':'count'})
list_df = [school,signal,cbd,other]
df = pd.concat([y18,school,signal,cbd,other,total]).rename(columns={'CREW_ASSIGNED':'Total Counts','BARS':'Total Bars','SHAPE_LEN':'Total Length'}).fillna('N/A')
#final = df.rename(columns=['Location','Bars','Length','Total Counts'])
display(school)
display(df)
writer = pd.ExcelWriter(r"C:\Users\Govs\CrosswalkData2019.xlsx")
df.to_excel(writer,'Total')
writer.save()
```
