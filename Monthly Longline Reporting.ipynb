```python
from arcgis.gis import GIS
from arcgis.features import *
import pandas as pd
import datetime
from arcgis.geometry import Geometry
```

```python
def createSDF(title):
    search= gis.content.search(query="title:" + title,item_type="Feature Layer")
    item = search[0]
    layer = item.layers[0]
    sdf = SpatialDataFrame.from_layer(layer)
    return sdf
```

```python
gis = GIS("http://austin.maps.arcgis.com/home/index.html")
longline = createSDF("Pavement Markings Work Orders")
longline["COMPLETED_DATE"] = pd.to_datetime(longline["COMPLETED_DATE"])
#longline.set_index(pd.DatetimeIndex(longline['COMPLETED_DATE'])) 
m_street = createSDF("atd_maintained_streets")

# Create clean filtered table
table = longline.filter(items=["ATD_WORK_ORDER_ID","WORKGROUP","REQUESTOR","COMPLETED_DATE","Shape__Length","SHAPE"]).query("WORKGROUP == 'LONG LINE'")
table['Month'],table['Day'],table['Year'] = table['COMPLETED_DATE'].dt.month,table['COMPLETED_DATE'].dt.day,table['COMPLETED_DATE'].dt.year
table = table.query("Month == 2 and Year == 2019")
```
