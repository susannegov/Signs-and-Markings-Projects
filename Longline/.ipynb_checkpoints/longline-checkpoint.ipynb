```python
import pandas as pd
```

```python
# First we only filter if past October 2020 and remove all letters
df = pd.read_csv("longline_jobs.csv")
df["Completed Date"] = pd.to_datetime(df["Completed Date"])
df['Street Segment IDs'] = df['Street Segment IDs'].replace(to_replace=['\(','\)','[A-Z]'], value="", regex=True)
df['Street Segment IDs'] = df['Street Segment IDs'].str.replace("\n", ',')
longline = df[df['Completed Date'] >= '2020-10-01']

# filter null street segment Ids column
longline = longline.dropna() 

#remove cap letters and parentheses
display(longline)
longline.to_csv("longline_clean.csv")
```
