## Methods
These functions will be used to extract and transform the data into a feasible format.

<i>The table below explains the purpose of each:</i>

| Method | Description   |
|:--------:|----|
|<b>df_text</b> |Changes element text to the pandas dataframe|
|<b>cover_aerial</b> |Zooms to selected features and extent for the dataframe|

```python
# Formats element text to match pandas dataframe
def df_text(r):
    date = datetime.datetime.now()
    for e in arcpy.mapping.ListLayoutElements(mxd,'TEXT_ELEMENT'):
            if e.name == 'SPECIFICATIONS':
                e.text = str(r["Specifications"])
            elif e.name == 'CREATED DATE':
                e.text =  "{}/{}/{}".format(date.month,date.day,date.year)
            elif e.name == 'REQUESTOR ID':
                e.text = str(r["Location ID"])
            elif e.name == 'LOCATION':
                e.text = "{} from {} to {}".format(r["Street"],r["From"],r["To"])
            elif e.name == 'author':
                e.text = 'Susanne Gov'
            elif e.name == 'WORK GROUPS':
                e.text = str(r['WORK GROUPS'])

# Refreshes dataframe map aerial                
def cover_aerial(mapdoc):
    dataframe = arcpy.mapping.ListDataFrames(mapdoc,"Layers")[0]
    dataframe.zoomToSelectedFeatures()
    dataframe.extent = layer.getSelectedExtent()
    arcpy.RefreshActiveView()
    
# Rename panda columns based on markings assets domain codes
def rename_markings_col(df):
    renameList = list(df.columns)
    arrow = ["Through","Left ","Right","Left and Right","Left, Right and Through","Left and Through", "Right and Through","U-turn","Lane reduction", "Wrong way","Bike"]
    other = ["Green pad", "Green launch pad", "Speed hump marking","Diagonal crosshatch", "Chevron crosshatch"]
    parking = ["Parking 'L'", "Parking 'T'", "Parking stall line", "Handicap symbol"]
    symbol = ["Bicycle (Bike)","Shared lane (Sharrow)", "Bicyclist","Railroad Crossing (RxR)", "Chevron","Pedestrian", "Diamond"]
    word = ["Stop", "Yield", "Ahead", "Only", "Merge","Ped", "X-ing","Bus Only", "Keep Clear", "Do Not Block", "Ped X-ing"]
    t =['word','arrow','symbol','','parking']
    st = [word,arrow,symbol,other,parking]
    index = 0
    for i in renameList:
        if isinstance(i,tuple):
            x = map(int,list(i))
            temp = st[x[0] - 1][x[1] - 1] + " " + t[x[0] - 1]
            renameList[index] = temp
        elif i == 'STOP_LINE' or i == 'CROSSWALK':
            temp = i.capitalize().replace('_','')
            renameList[index] = temp
        index += 1
    df.columns = renameList
    return df
```

## Import Datasets

```python
if Path(FILE_NAME + '.xlsx').exists():
    df = pd.read_excel(FILE_NAME + '.xlsx', index_col=0)
    df1 = pd.read_excel(FILE_NAME + '.xlsx','markings list', index_col=0).reset_index()

    sl = df1.groupby(['LOCATION ID','SHORTLINE TYPE']).count()[['SEGMENT ID']].rename(columns={"SEGMENT ID":'COUNTS'})
    sl = sl.pivot_table(values='COUNTS',index='LOCATION ID',columns='SHORTLINE TYPE',aggfunc='first').reset_index()
    sl['x'] = 'SHORT LINE'
    
    sp = df1.groupby(['LOCATION ID','SPECIALTY TYPE','SPECIALTY SUBTYPE']).count()[['SEGMENT ID']].rename(columns={"SEGMENT ID":'COUNTS'})
    sp = sp.pivot_table(values='COUNTS',index='LOCATION ID',columns=('SPECIALTY TYPE','SPECIALTY SUBTYPE'),aggfunc='first').reset_index()
    sp['y'] = 'SPECIALTY MARKINGS'
    
    df.loc[df.Comments.isnull() == False, 'll'] = 'LONGLINE'
    

    df = reduce(lambda z,y: pd.merge_ordered(z,y,left_on='Location ID', right_on ='LOCATION ID'), [df,sl,sp])
    df = df.dropna(how='all',subset=list(df.columns)[5:])
    df['WORK GROUPS'] = df[['x','ll',('y','')]].astype(str).apply(','.join,1)
    df['WORK GROUPS'] = df['WORK GROUPS'].apply(lambda x: [s for s in x.split(',') if s != 'nan'])
    df = df.drop(columns=['x','ll',('LOCATION ID',''),'LOCATION ID',('y','')]).dropna(1,'all').fillna('')
    df["Specifications"] = ''
    df = rename_markings_col(df).reset_index(drop=True)

    for index,row in df.iterrows():
        keys = list(row[6:])
        values = list(df.columns)[6:]
        spec = []
        for k,v in zip(keys,values):
            if k != '' and v != 'WORK GROUPS':
                spec.append('{} {}'.format(int(k),v.lower()))
        sentence = 'Install {}'.format(row['Comments']) + ', '.join(word for word in spec)
        df.at[index,'Specifications'] = sentence
    display(df)
```
